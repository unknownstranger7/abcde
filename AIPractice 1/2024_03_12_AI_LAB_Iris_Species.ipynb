```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("/content/drive/MyDrive/iris_species/Iris.csv")
data.head()
```

```python

X = data.drop('Species',axis=1).drop('Id',axis=1) 
Y = data['Species']
X.shape[1]
```

```python
data.describe()
```

```python
X.describe()
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
end=LabelEncoder()

Y=end.fit_transform(Y)
```

```python
Y
```

```python

Y = pd.get_dummies(Y).values
```

```python

Y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.activations import relu
from keras.optimizers import Adam
```

```python
model=Sequential()
model.add(Dense(20, relu, input_dim=X.shape[1])) #
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.summary()
```

```python

model.compile(loss='categorical_crossentropy', optimizer=Adam(0.01), metrics=['accuracy'])
```

here we have not divided the data into three parts, train test and , so we are using validation_split to split the data, here the training data in training and validation (0.2 -> so from 100% training data , 20% is used for validation and 80% for training)

```python
hist=model.fit(xtrain, ytrain, epochs=50, batch_size=32, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(hist.history['loss'], label='loss')
plt.plot(hist.history['val_loss'], label='val_loss')
plt.xlabel('no. of iterations')
plt.ylabel("Loss")
plt.legend()
```

```python
plt.plot(hist.history['accuracy'], label='accuracy')
plt.plot(hist.history['val_accuracy'], label='val_accuracy')
plt.xlabel('no. of iterations')
plt.ylabel("Accuracy")
plt.legend()
```

```python
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
actual = np.argmax(ypred, axis=1)
```

```python
pred = np.argmax(ytest, axis=1)
```

```python
print(classification_report(actual,pred))
```
